#### Handling missing values


```python
# These columns have + symbols. we need to remove that symbol
# Split the column and remove the '+' with the values after them
columns=[ 'ls', 'st', 'rs', 'lw', 'lf', 'cf', 'rf', 
         'rw', 'lam', 'cam', 'ram', 'lm', 'lcm', 'cm',
         'rcm', 'rm', 'lwb', 'ldm', 'cdm', 'rdm', 'rwb', 'lb', 
         'lcb', 'cb','rcb', 'rb']
data[columns]

for column in columns:
    data[column]= data[column].str.split('+',expand=True)[0]
data.head(5)
```

```python
mis_val=['ls', 'st', 'rs', 'lw', 'lf', 'cf', 'rf', 
         'rw', 'lam', 'cam', 'ram', 'lm', 'lcm',
         'cm', 'rcm', 'rm', 'lwb', 'ldm', 'cdm', 
         'rdm', 'rwb', 'lb', 'lcb', 'cb', 'rcb', 'rb']
for x in mis_val:
    med_val = data[x].median()
    data[x].fillna(med_val,inplace=True)
```

```python
data.isnull().sum()
```

```python
Nan_count=data.isnull().sum()
drop_col= Nan_count[Nan_count >10000].index
data.drop(columns=drop_col,inplace=True)
```

## Model Evalution:

### PRINCIPLE COMPONENT ANALYSIS(PCA)

```python
# Step:1 Getting optimal value of pca
from sklearn.decomposition import PCA
pca = PCA() 
principlecomponents = pca.fit_transform(data) # Fitting the data
plt.figure(figsize=(10,5))
plt.plot(np.cumsum(pca.explained_variance_ratio_),marker='*',color='k') #plot the variance ratio
plt.xlabel('Number of components')
plt.ylabel('Variance(%)') # for each componant
plt.title('Explained Variance',fontsize=20)
plt.show()
```

```python

# Step:2 Select the components
pca = PCA(n_components=7)
fifa = pca.fit_transform(data)

# Step:3 Make new dataframe
pca_data= pd.DataFrame(data=fifa,columns=['pca1','pca2','pca3','pca4','pca5','pca6','pca7'])
print(pca.explained_variance_ratio_)
pca_data.head()
```

## K-MEANS CLUSTERING


```python
X = pca_data
X.head()
```

```python
# Compute eigenvalues
eigenvalues = pca.explained_variance_

# Print eigenvalues
print("Eigenvalues:")
print(eigenvalues)
```

* The first eigenvalue (1.13848200e+14) is much larger than the rest. This suggests that the first principal component explains a significant amount of variance in the data.

#### Elbow method is the most popular method that is used to determine the optimal value of k.

```python
#elbow method
from sklearn.cluster import KMeans
wcss=[]
for i in range (1,11):
    kmeans=KMeans(n_clusters=i,random_state=42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
plt.figure(figsize=(10,4))
plt.plot(range(1,11),wcss,ls="--",marker="o",color="b")
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
#We will take the optimal value of k as 3
model = KMeans(n_clusters=3, random_state=42)

#Training the model
model.fit(X)
```

```python
# getting labels
label=model.labels_
label

# cluster centroids
model.cluster_centers_
```

### Evaluation metrics

### Silhouette Score:
The range of Silhouette score is [-1, 1]. Its analysis is as follows −

* +1 Score − Near +1 Silhouette score indicates that the sample is far away from its neighboring cluster.

* 0 Score − 0 Silhouette score indicates that the sample is on or very close to the decision boundary separating two neighboring clusters.

* -1 Score − 1 Silhouette score indicates that the samples have been assigned to the wrong clusters.

```python
from sklearn.metrics import silhouette_score
score=silhouette_score(X,label)
print("Silhoutte score for 3 clusters:",score)
```

```python
#Checking clustering using k ass 5
#Now,We will take the optimal value of k as 5
model1 = KMeans(n_clusters=5, random_state=42)

#Training the model
model1.fit(X)
```

```python
#gtting labels for 5 clusters
label1= model1.labels_
label1
 
#5 cluster centroids
model1.cluster_centers_
```

```python
# Evaluation metrics for 5 clusters
from sklearn.metrics import silhouette_score
score=silhouette_score(X,label1)
print("Silhoutte score for 5 clusters:",score)
```
